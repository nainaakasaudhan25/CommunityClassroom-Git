```python
import pandas as pd
df = pd.read_csv("ai_mock_interview_labeled_dataset_30000_rows.csv")
```

```python
print(df.shape)
print(df.columns)
print(df["label"].value_counts())
print(df.isnull().sum())
```

```python
df["text"]=df["question"] + " " + df["answer"]
```

```python
df = df.drop_duplicates(subset=["text"])
print(df.shape)
```

```python
from sklearn.model_selection import train_test_split

train_texts, val_texts, train_labels, val_labels = train_test_split(
    df["text"].tolist(),
    df["label"].tolist(),
    test_size=0.2,
    random_state=42,
    stratify=df["label"]
)
```

```python
print(len(set(train_texts).intersection(set(val_texts))))
```

```python
print(len(train_texts))
print(len(val_texts))
print(len(train_labels))
print(len(val_labels))
```

```python
from transformers import BertTokenizer
tokenizer=BertTokenizer.from_pretrained("bert-base-uncased")
```

```python
train_encodings = tokenizer(
    train_texts,
    truncation=True,
    padding=True,
    max_length=128
)

val_encodings = tokenizer(
    val_texts,
    truncation=True,
    padding=True,
    max_length=128
)
```

```python
import torch

class InterviewDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item["labels"] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

train_dataset = InterviewDataset(train_encodings, train_labels)
val_dataset = InterviewDataset(val_encodings, val_labels)
```

```python
from transformers import BertForSequenceClassification

model = BertForSequenceClassification.from_pretrained(
    "bert-base-uncased",
    num_labels=3
)
```

```python
import torch
torch.cuda.is_available()
```

```python
#!nvidia-smi
```

```python
from transformers import Trainer, TrainingArguments

training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=2,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    do_train=True,
    do_eval=True,
    fp16=False
)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset
)
```

```python
trainer.evaluate()
```

```python
trainer.train()
```
